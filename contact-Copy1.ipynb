```python
def update_contact(a):
    
    new = input('Enter new Name: ').title()
    contact_data.loc[contact_data[contact_data.Name== a].index.values,'Name']=new

    new1 = int(input('Enter new Number: '))
    contact_data.loc[contact_data[contact_data.Name== new].index.values,'Number']=new1

    print('Update  Number 2 Y/N')
    want_num = input().title()
    if want_num == 'Y':
        new12 = int(input('Enter new Number: '))
        contact_data.loc[contact_data[contact_data.Name==new].index.values,'Number1']=new12
    else:
        pass

    print('Update  Number 3 Y/N')
    want_num1 = input().title()
    if want_num1 == 'Y':
        new13 = int(input('Enter new Number: '))
        contact_data.loc[contact_data[contact_data.Name==new].index.values,'Number2']=new13
    else:
        pass

    new2 = input('Enter new Email: ')
    contact_data.loc[contact_data[contact_data.Name== new].index.values,'Email']=new2

    new3 = input('Enter new NickName: ').title()
    contact_data.loc[contact_data[contact_data.Name== new].index.values,'NickName']=new3
    print('\nDone Successful','\n',contact_data)
```

# Main code

```python
def Contact_Book():
    flag = 4
    print('''
    .....Welcome to Contact Book Mangement....
    ...........To Open Press Y/N..............''')
    open1 = input('Y/N: ').title()
    if open1 == 'Y':
        while flag:
            print('''
            To Open Contact Book                 Press 1
            To Add New Contact                   Press 2
            To Search Contact                    Press 3
            To Update Name/Number/Email/NickName Press 4
            To Delete Contact                    Press 5
            To Exit                              Press 0 ''')
            Option1 = int(input('Option: '))
            if Option1 == 1:
                print(contact_data)

            elif Option1 == 2:
                add_details()

            elif Option1 == 3:
                search_data()

            elif Option1 == 4:
                update_data()

            elif Option1 == 5:
                delete_data()

            elif Option1== 0:
                print('/nThank You For Visiting \n....Contact Book.....')
                break
            else:
                print('\n.....Invalid input.....')
                print(f'You have {flag - 1} Chances Left....')
                flag-=1      

    elif open1 == 'N':
        print('/nThank You For Visiting \n....Contact Book.....')
    else:
        print('Invalid Input')
        
```
